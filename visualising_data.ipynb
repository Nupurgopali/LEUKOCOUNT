#### Convert EOSINOPHIL cell image to gray scale and then process the image further

```python
def grey_image(image,cell_name):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    plt.imshow(gray)
    plt.title('GRAY SCALE {}'.format(cell_name)) 
    return gray
```

```python
grey=grey_image(im,'EOSINOPHIL')
```

#### Thresholding is used to compare pixel values with a threshold value and updates it accordingly. A simple threshold will set pixel values to 1 if there value is more than the threshold and 0 if the value is less than threshold.

#### In case of adaptive thresholding, different threshold values are used for different parts of the image. Here I used adaptive thresholding to get better idea about the image.

```python
def threshold(gray_image):
    th2 = cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    plt.imshow(th2)
    return th2
```

```python
th2 = threshold(gray)
```

#### Edges are the points in an image where the image brightness changes sharply or has discontinuities.Edges are very useful features of an image that can be used for different applications like classification of objects in the image and localization.Edges are different from contours as they are not related to objects rather they signify the changes in pixel values of an image. Edge detection can be used for image segmentation and even for image sharpening.


```python
def edge_detection(image):
    edges = cv2.Canny(im,20,100)
    plt.imshow(edges)
    return edges
```

```python
edges=edge_detection(im)
```
